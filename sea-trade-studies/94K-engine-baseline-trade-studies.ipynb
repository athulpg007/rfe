```python
output_file1 = 'data\output-1D-sweep-gamma-baseline-u1'
param_arr1  = np.loadtxt(output_file1+'-param.dat')
powero_arr1 = np.loadtxt(output_file1+'-pow.dat')
heatin_arr1 = np.loadtxt(output_file1+'-qin.dat')
effact_arr1 = np.loadtxt(output_file1+'-eff.dat')

fig = plt.figure(figsize=(8,6))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.plot(param_arr1/1e-6, effact_arr1, color='r', marker='s', mfc='none', 
          markeredgewidth=3, markersize=6, linestyle='solid', label=r'Efficiency')
         
plt.xlabel('Compression space swept volume, cc',fontsize=16)
plt.ylabel('Thermal Efficiency', fontsize=16, color='red')

plt.title(r'$T_C=273 K, T_H=900K, 10-bar $'+', He, ' + r'$\gamma$-engine' , fontsize=16)
         
ax = plt.gca()
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)

ax.xaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.xaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax.yaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.yaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax.xaxis.set_tick_params(length=5, width=2)
ax.yaxis.set_tick_params(length=5, width=2, colors='r')

ax.tick_params(direction='in')
#ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=16)
ax.tick_params(axis='y',labelsize=16)

ax.spines['left'].set_color('r')
ax.spines['left'].set_linewidth(2)
ax.legend(loc='lower center', fontsize=16)
ax.set_ylim([0,1.1*max(effact_arr1)])

ax2 = ax.twinx()
ax2.plot(param_arr1/1e-6, powero_arr1, color='b', marker='o', mfc='none', 
         markeredgewidth=3, markersize=6, linestyle='solid', label=r'Power')

ax2.xaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax2.xaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax2.yaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax2.yaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax2.xaxis.set_tick_params(length=5, width=2)
ax2.yaxis.set_tick_params(length=5, width=2)

ax2.tick_params(direction='in')
#ax2.yaxis.set_ticks_position('both')
ax2.xaxis.set_ticks_position('both')
ax2.tick_params(axis='x',labelsize=16)
ax2.tick_params(axis='y',labelsize=16, colors='b')

ax2.set_xlabel('Expansion Phase Angle Advance',fontsize=16)
ax2.set_ylabel('Output power, W', fontsize=16, color='b')
ax2.spines['right'].set_color('b')
ax2.spines['right'].set_linewidth(2)
ax2.legend(loc='upper right', fontsize=16)
ax2.set_ylim([0,1.1*max( powero_arr1)])

plt.savefig('plots/gamma-baseline-u1.png', dpi=300, bbox_inches='tight')
plt.savefig('plots/gamma-baseline-u1.pdf', dpi=300,bbox_inches='tight')
plt.savefig('plots/gamma-baseline-u1.eps', dpi=300,bbox_inches='tight')
```

```python
from scipy import interpolate
```

```python
output_file1 = 'data/output-2D-sweep-gamma-baseline-u20'

param1_arr  = np.loadtxt(output_file1+'-param1.dat')
param2_arr  = np.loadtxt(output_file1+'-param2.dat')

powero_arr1 = np.loadtxt(output_file1+'-pow.dat')
heatin_arr1 = np.loadtxt(output_file1+'-qin.dat')
effact_arr1 = np.loadtxt(output_file1+'-eff.dat')

p1 = interpolate.interp2d(param1_arr, param2_arr, powero_arr1, kind='cubic')
q1 = interpolate.interp2d(param1_arr, param2_arr, heatin_arr1, kind='cubic')
e1 = interpolate.interp2d(param1_arr, param2_arr, effact_arr1, kind='cubic')

x_new =  np.linspace( param1_arr[0],  param1_arr[-1],  100)
y_new =  np.linspace( param2_arr[0],  param2_arr[-1],  100)

p1_new =  np.zeros((len(x_new),len(y_new)))
q1_new =  np.zeros((len(x_new),len(y_new)))
e1_new =  np.zeros((len(x_new),len(y_new)))

for i in range(0,len(x_new)):
    for j in range(0,len(y_new)):

        p1_new[i,j] = p1(x_new[i],y_new[j])
        q1_new[i,j] = q1(x_new[i],y_new[j])
        e1_new[i,j] = e1(x_new[i],y_new[j])
           
P1 = p1_new
Q1 = q1_new
E1 = e1_new

X, Y = np.meshgrid(x_new, y_new)

fig = plt.figure()

fig.set_size_inches([16,6])
plt.subplot(121)
plt.ion()
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)
plt.xlabel('Compression space, cc',fontsize=16)
plt.ylabel('Expansion space, cc', fontsize=16, color='k')

plt.title(r'$T_C=273 K, T_H=900K, 10-bar $'+', He, ' + r'$\gamma$-engine' , fontsize=16)
         
ax = plt.gca()
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)

ax.xaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.xaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax.yaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.yaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax.xaxis.set_tick_params(length=5, width=2)
ax.yaxis.set_tick_params(length=5, width=2, colors='k')

ax.tick_params(direction='in')
#ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=16)
ax.tick_params(axis='y',labelsize=16)

ax.spines['left'].set_color('k')
ax.spines['left'].set_linewidth(2)
#ax.legend(loc='lower center', fontsize=16)

plt.contourf(X/1e-6, Y/1e-6, E1, 100, cmap='jet')
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=16)
cbar.set_label(r'Efficiency', labelpad=-27, y=1.07, rotation=0, fontsize=16)
cbar.ax.tick_params(axis='y', direction='in')

CS1 = plt.contour(X/1e-6, Y/1e-6, E1, levels=20)
ax.clabel(CS1, CS1.levels, inline=True, colors = 'k', fontsize=12)

plt.subplot(122)
plt.xlabel('Compression space, cc',fontsize=16)
plt.ylabel('Expansion space, cc', fontsize=16, color='k')

plt.title(r'$T_C=273 K, T_H=900K, 10-bar $'+', He, ' + r'$\gamma$-engine' , fontsize=16)
         
ax = plt.gca()
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)

ax.xaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.xaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax.yaxis.grid(which='major', color='k', linestyle='dotted', linewidth=0.5)
ax.yaxis.grid(which='minor', color='k', linestyle='dotted', linewidth=0.5)

ax.xaxis.set_tick_params(length=5, width=2)
ax.yaxis.set_tick_params(length=5, width=2, colors='k')

ax.tick_params(direction='in')
#ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=16)
ax.tick_params(axis='y',labelsize=16)

ax.spines['left'].set_color('k')
ax.spines['left'].set_linewidth(2)
#ax.legend(loc='lower center', fontsize=16)

plt.contourf(X/1e-6, Y/1e-6, P1, 100, cmap='jet')
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=16)
cbar.set_label(r'Power, W', labelpad=-27, y=1.07, rotation=0, fontsize=16)
cbar.ax.tick_params(axis='y', direction='in')

CS2 = plt.contour(X/1e-6, Y/1e-6, P1, levels=20)
ax.clabel(CS2, CS2.levels, inline=True, fmt= "%3d", colors = 'k', fontsize=12)

plt.savefig('plots/gamma-baseline-u20.png', dpi=300, bbox_inches='tight')
plt.savefig('plots/gamma-baseline-u20.pdf', dpi=300,bbox_inches='tight')
plt.savefig('plots/gamma-baseline-u20.eps', dpi=300,bbox_inches='tight')
```
